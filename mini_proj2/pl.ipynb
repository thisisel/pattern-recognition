```python
import matplotlib.pyplot as plt

# Data
num_components = [1, 2, 3]
pca_accuracies = [90.63, 85.94, 93.31]
flda_accuracies = [96.64, 88.67, 79.39]

# Create the plot
plt.figure(figsize=(8, 4))
plt.plot(num_components, pca_accuracies, label='PCA', marker='o')
plt.plot(num_components, flda_accuracies, label='FLDA', marker='o')
plt.xlabel('Number of Principal Components')
plt.ylabel('% Average Accuracy')
plt.title('Accuracy vs. Number of Principal Components')
plt.xticks(num_components)
plt.legend()
plt.grid(True)
plt.show()
```
